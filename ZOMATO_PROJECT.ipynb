# ⏩ **COPY & CLEAN DATASET**

```python
# copy dataset for the work
df1 = zomato.copy()
df1.head()
```

```python
# Remove some unwanted columns

df1.drop(columns=["url","phone","address","menu_item"],axis=1,inplace=True)
df1.head()
```

```python
# clean the "rate" columns

def handel(value):
  if (value == 'NEW') or (value == '-'):
    return np.nan
  else:
    value = str(value).split('/')
    value = value[0]
    return float(value)

df1["rate"] = df1["rate"].apply(handel) # Assign the result back to the 'rate' column
```

```python
df1.head()
```

# ⏩ **REMOVE DUBPLICATES AND NULL VALUE**

```python
df1.duplicated().sum()
```

```python
df1.drop_duplicates(inplace=True)
df1.duplicated().sum()
```

```python
df1.isnull().sum()
```

```python
# [Step:1] "Rate columns"
# There are "10052" value are missing in "Rate columns"
# so we can find "mean" "middle" "mode" and after we can fill suitable value
```

```python
df1["rate"].mean()
```

```python
df1["rate"].median()
```

```python
df1["rate"].mode()
```

```python
"""
conclusion:
--> We can see the "mean" "middle" "mode" and we can fill "3.7" value in null value
"""
```

```python
df1["rate"].fillna(3.7,inplace=True)
```

```python
df1.isnull().sum()
```

# ⏬ 4. What was the minimum maximum and average rating of data set

```python
plt.hist(df1['rate'], bins=10, edgecolor='black')
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')

# Calculate and display minimum, maximum, and average
min_rate = df1['rate'].min()
max_rate = df1['rate'].max()
avg_rate = df1['rate'].mean()

plt.text(0.05, 0.95, f'Min: {min_rate:.2f}', transform=plt.gca().transAxes)
plt.text(0.05, 0.90, f'Max: {max_rate:.2f}', transform=plt.gca().transAxes)
plt.text(0.05, 0.85, f'Avg: {avg_rate:.2f}', transform=plt.gca().transAxes)

plt.show()
```

```python
"""
conclusion:

This histogram represents the distribution of restaurant ratings. Key conclusions are:

-->Most ratings are between 3.5 and 4.0:
The majority of restaurants fall within this range, indicating generally positive customer feedback.

--> Normal distribution:
The ratings show a bell-shaped curve, suggesting that the distribution is fairly normal, with fewer restaurants rated very low or very high.

--> Average rating is 3.7:
This aligns with the peak of the distribution, indicating that most restaurants perform around this level.

--> Outliers exist:
Some restaurants have ratings as low as 1.8 and as high as 4.9, but these are rare.

Overall, most restaurants are rated favorably, though exceptional and poorly rated establishments are less common.
"""
```

# ⏬ 6. Restaurants allowing table booking or not ?

```python
# assign data
booking_counts = df1["book_table"].value_counts()
colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(booking_counts.values, labels=booking_counts.index, explode=explode, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Restaurants Allowing Table Booking')
plt.axis('equal')
plt.show()
```

```python
"""
**Conclusion:**

This pie chart depicts the proportion of restaurants that allow table booking. Key insights are:

--> **Limited table booking availability:** Only 12.6% of restaurants provide table booking facilities, indicating that the majority do not support this service.

--> **Dominance of non-booking restaurants:** A significant 87.4% of restaurants do not allow table reservations, highlighting a preference for walk-in or casual dining setups.

--> **Niche table booking market:** The small percentage of restaurants offering bookings suggests it caters to specific customer segments seeking convenience or premium services.

In conclusion, table booking is not a widely available feature in restaurants, with most establishments focusing on non-reservation-based dining.
"""
```

# ⏬ 8. Relation between Location and Rating.

```python
import matplotlib.pyplot as plt

# Group data by location and calculate average rating
location_ratings = df1.groupby('location')['rate'].mean().reset_index()

# Create scatter plot
plt.figure(figsize=(35, 6))
plt.scatter(location_ratings['location'], location_ratings['rate'])
plt.title('Location vs. Rating Scatter Plot')
plt.xlabel('Location')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
"""
Conclusion:

This scatter plot illustrates the average restaurant ratings across different locations. Key insights are:

--> Consistency in ratings:
Most locations have an average rating between 3.4 and 3.8, indicating a stable level of customer satisfaction.

--> High-rated outliers:
A few locations stand out with ratings exceeding 4.0, showcasing exceptional dining experiences in these areas.

--> Lower-rated areas:
Some locations fall closer to the lower end of the spectrum, around 3.2, highlighting potential areas for improvement.

--> Even distribution:
The ratings are fairly evenly spread across locations, reflecting diverse customer experiences in different areas.

In conclusion, while the majority of locations maintain decent average ratings, a few areas achieve excellence, and others have room for growth.
"""
```
